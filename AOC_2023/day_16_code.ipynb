```python
#import script1
import math

class Solution:
    def findMedianSortedArrays(array1: list, array2: list) -> float:
        mergedArray = []
        i = 0
        j = 0
        i_max = len(array1)
        j_max = len(array2)

        while (i < i_max) or (j < j_max):
            if i == i_max:
                mergedArray.append(array2[j])
                j += 1
                continue
            if j == j_max:
                mergedArray.append(array1[i])
                i += 1
                continue

            e1 = array1[i]
            e2 = array2[j]

            if e1 < e2:
                mergedArray.append(e1)
                i += 1
            else:
                mergedArray.append(e2)
                j += 1

        m = (i_max + j_max - 1)/2
        m_1, m_2 = math.floor(m), math.ceil(m)

        return (mergedArray[m_1] + mergedArray[m_2])/2

testCase = readNumsFromInput()
print(
    Solution.findMedianSortedArrays(testCase[0], testCase[1])
)

```
